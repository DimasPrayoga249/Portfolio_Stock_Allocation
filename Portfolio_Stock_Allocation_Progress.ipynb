```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import dateutil.relativedelta
import warnings
import pandas as pd
```

```python
# Mengabaikan FutureWarning dari yfinance
warnings.filterwarnings('ignore', category=FutureWarning, module='yfinance')

import mplfinance as mpf # Import mplfinance untuk candlestick chart

def plot_stock_ranges(ticker):
    """
    Mengambil data historis saham dari Yahoo Finance untuk berbagai rentang waktu
    dan memplot pergerakan harga penutupan.

    Args:
        ticker (str): Simbol ticker saham (misalnya 'MSFT', 'AMZN', 'META').
    """
    # Definisikan rentang waktu akhir (hari ini)
    end_date_obj = datetime.today()
    end_date = end_date_obj.strftime('%Y-%m-%d')

    # Mapping dari string rentang ke objek relativedelta untuk perhitungan yang akurat
    range_map = {
        "1 week": dateutil.relativedelta.relativedelta(weeks=1),
        "1 month": dateutil.relativedelta.relativedelta(months=1),
        "1 quarter": dateutil.relativedelta.relativedelta(months=3), # 1 kuartal = 3 bulan
        "1 year": dateutil.relativedelta.relativedelta(years=1),
        "5 years": dateutil.relativedelta.relativedelta(years=5),
        "10 years": dateutil.relativedelta.relativedelta(years=10),
        "alltime": "max", # Menggunakan string 'max' untuk periode yfinance
    }

    # Loop melalui setiap rentang waktu dan plot grafik
    for selected_range_str, delta_obj in range_map.items():
        # Handle 'alltime' case separately
        if selected_range_str == "alltime":
            print(f"\nMengambil data untuk {ticker} dalam rentang {selected_range_str} (dari awal hingga {end_date})...")
            data = yf.download(ticker, period='max', auto_adjust=True)
            start_date = data.index.min().strftime('%Y-%m-%d') # Get actual start date for title
        else:
            start_date_obj = end_date_obj - delta_obj
            start_date = start_date_obj.strftime('%Y-%m-%d')
            print(f"\nMengambil data untuk {ticker} dalam rentang {selected_range_str} ({start_date} hingga {end_date})...")
            data = yf.download(ticker, start=start_date, end=end_date, auto_adjust=True)

        # Pastikan ada data yang diambil sebelum plotting
        if not data.empty:
            plt.figure(figsize=(10, 5))
            plt.plot(data['Close'])
            plt.title(f'Pergerakan Harga Saham {ticker} ({selected_range_str}) dari {start_date} hingga {end_date}')
            plt.xlabel('Tanggal')
            plt.ylabel('Harga Penutupan (USD)')
            plt.grid(True)
            plt.show()
        else:
            print(f"Tidak ada data yang tersedia untuk {ticker} dalam rentang {selected_range_str}.")
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime
import dateutil.relativedelta

def plot_growth_chart():

    tickers = ["MSFT", "AMZN", "META", "LLY", "V"]

    end_date_obj = datetime.today()
    end_date = end_date_obj.strftime('%Y-%m-%d')

    range_map = {
        "1 Week": dateutil.relativedelta.relativedelta(weeks=1),
        "1 Month": dateutil.relativedelta.relativedelta(months=1),
        "1 Quarter": dateutil.relativedelta.relativedelta(months=3),
        "1 Year": dateutil.relativedelta.relativedelta(years=1),
        "5 Years": dateutil.relativedelta.relativedelta(years=5),
        "10 Years": dateutil.relativedelta.relativedelta(years=10)
    }

    for range_name, delta_obj in range_map.items():

        percent_changes = []

        for ticker in tickers:

            if range_name == "All Time":
                data = yf.download(ticker, period="max", auto_adjust=True)
            else:
                start_date_obj = end_date_obj - delta_obj
                start_date = start_date_obj.strftime('%Y-%m-%d')
                data = yf.download(ticker, start=start_date, end=end_date, auto_adjust=True)

            if not data.empty:

                close_prices = data["Close"]

                # Handle jika MultiIndex
                if isinstance(close_prices, pd.DataFrame):
                    close_prices = close_prices.iloc[:, 0]

                start_price = float(close_prices.iloc[0])
                end_price = float(close_prices.iloc[-1])

                percent_change = ((end_price - start_price) / start_price) * 100
                percent_changes.append(percent_change)

            else:
                percent_changes.append(0)

        # Plot 1 chart untuk tiap range
        plt.figure(figsize=(8, 4))
        plt.bar(tickers, percent_changes)
        plt.title(f"Persentase Kenaikan / Penurunan Harga Emiten ({range_name})")
        plt.ylabel("Percent Change (%)")
        plt.xlabel("Ticker")
        plt.axhline(0)
        plt.show()
```

```python
plot_growth_chart()
```

```python
def calculate_portfolio_value(portfolio):
    total_value = 0
    results = {}

    # Hitung nilai masing-masing saham
    for ticker, shares in portfolio.items():
        data = yf.download(ticker, period="1d", auto_adjust=True)

        if not data.empty:

          close_prices = data["Close"]

          # Jika masih DataFrame (MultiIndex)
          if isinstance(close_prices, pd.DataFrame):
              close_prices = close_prices.squeeze()

          # Jika masih Series → ambil elemen terakhir
          if isinstance(close_prices, pd.Series):
              current_price = close_prices.iloc[-1]
          else:
              # Jika sudah scalar
              current_price = close_prices

              current_price = float(current_price)

              value = shares * current_price

              results[ticker] = {
                  "Shares": shares,
                  "Current Price": round(current_price, 2),
                  "Total Value": value
              }

              total_value += value

    # Hitung persentase kontribusi
    for ticker in results:
        percentage = (results[ticker]["Total Value"] / total_value) * 100
        results[ticker]["Portfolio %"] = round(percentage, 2)
        results[ticker]["Total Value"] = round(results[ticker]["Total Value"], 2)

    # Tampilkan hasil
    print("\nNilai Portofolio:")
    for ticker, info in results.items():
        print(f"{ticker}: ${info['Total Value']} ({info['Portfolio %']}%)")

    print(f"\nTotal Nilai Portofolio: ${round(total_value, 2)}\n")
```

```python
calculate_portfolio_value(shares_portfolio)
```
